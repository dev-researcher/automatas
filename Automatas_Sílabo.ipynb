#### Descripción del Curso
Este curso proporciona una introducción comprensiva a la teoría de autómatas utilizando el libro "Automata Theory: An Algorithmic Approach" de Javier Esparza y Michael Blondin como texto guía. Los autómatas serán estudiados como estructuras de datos, con aplicaciones en verificación de programas, lógica y aritmética de Presburger. El curso combinará teoría con práctica utilizando Python en Google Colab.

#### Metodología de Clase
- **Primera hora:** Inducción y presentación de conceptos por el profesor.
- **Segunda hora:** Trabajo en grupos para resolver problemas prácticos.
- **Tercera hora:** Presentación de resultados y discusión liderada por los estudiantes.

El curso utilizará un libro complejo, por lo que cada estudainte deberá realizar lecturas asignadas.

### Evaluación

- **Participación**: 10%
- **Presentación Final:** 40%
